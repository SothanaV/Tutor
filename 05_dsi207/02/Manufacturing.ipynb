```python
import matplotlib.pyplot as plt
```

```python
base_a=[]
base_b=[]
for i in range(1000):
    base_a.append(i)
    base_b.append(0)
```

```python
base_a
```

```python
base_b
```

```python
plt.plot(base_a,base_b,color='black')
plt.plot(base_b,base_a,color='black')
plt.grid()
```

```python
x1 = []
y1 = []
for x in range(1000):
    y = (-(2/3)*x)+500
    y1.append(y)
    x1.append(x)
```

```python
plt.plot(x1,y1)
```

```python
x2 = []
y2 = []
for x in range(1000):
    y = (-2*x)+1000
    y2.append(y)
    x2.append(x)
```

```python
plt.plot(x2,y2)
```

```python
plt.plot(base_a,base_b,color='black')
plt.plot(base_b,base_a,color='black')
plt.grid()
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
# 2x + 3y = 1500
# จุดตัด (750,0) และ (0,500)
# x : 0 - 750
## -50 ถึง 800
x1 = []
y1 = []
for x in range(-50,800):
    y = (-(2/3)*x)+500
    y1.append(y)
    x1.append(x)
```

```python
# 2x + y = 1000
# จุดตัด (500,0) และ (0,1000)
# x : 0 - 500
## -50 ถึง 550
x2 = []
y2 = []
for x in range(-50,550):
    y = (-2*x)+1000
    y2.append(y)
    x2.append(x)
```

```python
plt.plot(base_a,base_b,color='black')
plt.plot(base_b,base_a,color='black')
plt.grid()
plt.plot(x1,y1,label='2x + 3y = 1500')
plt.plot(x2,y2)
plt.scatter(375,250,s=50,color='r')
plt.legend()
plt.title('manufacturing')
plt.xlabel('number of pants')
```
