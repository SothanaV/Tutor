```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
noOfRandomNumbers = 10000  # no of random integers
noOfHistogramElements = 20  # no of entries for the histogram
shift = noOfHistogramElements/2  # x axis shift for plot
```

```python
# 1a) Calculate the odds and log(odds) for 5:3
odds = 5 / 3
logOdds = np.log(odds)
print('5:3 odds: {:6.2f} logOdds: {:6.2f}'.format(odds, logOdds))
```

```python
# 1b) Plot the logit function
# p = np.arange(0.01, 0.99, 0.01)  # Alternative A for creation of x categories for histogram: arange
p = np.linspace(0.01, 0.99, 100)   # Alternative B: linspace
logit_p = np.log(p / (1 - p))  # log(odds) = logit(p)
plt.plot(p, logit_p)
plt.grid(b=True, which='major', axis='both')
plt.show()
```

```python
# 1c) Create random numbers that add up to 100, calc. the log(odds) and put the values in a histogram. What do you get?
histogramList = [0] * noOfHistogramElements
for i in range(0, noOfRandomNumbers): # create the random numbers, log(odds) values, histogram entries
    randomNumber = np.random.randint(1, 101)
    # np.random.randn(50) # standard normal dist.
    odds = randomNumber/(101-randomNumber) # calc. the odds for this event
    logOdds = np.log(odds)  # calc. the log(odds)
    logOddsInt = np.int_(round(logOdds + shift))  # convert to int in order to count for histogram
    print('randomNumber: {:6.2f} odds: {:6.2f} logOdds: {:6.2f} logOddsInt: {:6.2f} '.format(randomNumber, odds, logOdds, logOddsInt))
    histogramList[logOddsInt] += 1  # increment the count for the category with index logOddsInt
```

```python
# plot histogram entries
x = np.arange(-shift, +shift, 1)
plt.bar(x, histogramList)
# plt.xlim([-10, 10])
plt.ylabel('Odds')
plt.grid(b=True, which='major', axis='both')
plt.show()
```

```python
# 1d) calc the probability for throwing a die (with number of guesses as the independent variable)
outcomes = np.arange(1, 7, 1)
prob = []
for outcome in outcomes:
    prob.append(outcome/outcomes.size)  # for the last event 6 guesses the prob. is 1
```

```python
# calc the odds for throwing on die
odds = []
for outcome in outcomes:
    odds.append(outcome/(outcomes.size-outcome))  # for the last event 6 guesses the odds 6/(6-6)=6/0 -> Not defined = Error
```

```python
# plot the prob. and odds in a bar chart
plt.bar(outcomes-0.1, prob, width=0.3)
plt.bar(outcomes+0.1 , odds, width=0.3)
plt.xlim([0, 7])
plt.ylabel('Prob')
plt.grid(b=True, which='major', axis='both')
plt.show()
```
